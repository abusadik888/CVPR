```python
import matplotlib.pyplot as plt
import numpy as np
  
x = np.arange(-10, 11, 1)
print(x)
```

```python
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y,"STEP(X)")
```

```python
y = 1 / (1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "TanH(X)")
```

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y,"ReLU(X)")
```

```python
def elu(x, alpha=1.0):
    # Exponential Linear Unit activation function
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-10, 11, 1)
y = elu(x)

plt.figure()
plt.plot(x, y, 'o--')
plt.grid(True)
plt.xlabel("X")
plt.ylabel("ELU(X)")
plt.title("Exponential Linear Unit (ELU) Activation Function")
plt.show()
```

```python


def selu(x, alpha=1.67326, scale=1.0507):
    # Scaled Exponential Linear Unit activation function
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-10, 11, 1)
y = selu(x)

plt.figure()
plt.plot(x, y, 'o--')
plt.grid(True)
plt.xlabel("X")
plt.ylabel("SELU(X)")
plt.title("Scaled Exponential Linear Unit (SELU) Activation Function")
plt.show()
```
